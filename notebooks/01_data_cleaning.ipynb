## Feature Engineering

```python
df["NPK_Total"] = (
    df["N"] +
    df["P"] +
    df["K"]
)
```

```python
df["NPK_Average"] = (
    df["N"] +
    df["P"] +
    df["K"]
) / 3
```

```python
df["Humidity_Temp_Index"] = (
    df["humidity"] *
    df["temperature"]
)
```

```python
df["Soil_Fertility"] = (
    0.4 * df["N"] +
    0.3 * df["P"] +
    0.3 * df["K"]
)
```

```python
df["Rainfall_Category"] = pd.cut(
    df["rainfall"],
    bins=[0,100,200,400],
    labels=["Low","Medium","High"]
)
```
