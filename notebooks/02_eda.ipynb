```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

plt.style.use("ggplot")
```

```python
df = pd.read_csv("../data/cleaned_crop_data.csv")

df.head()
```

## Visualization 1

### Crop Distribution

```python
plt.figure(figsize=(12,6))
sns.countplot(
    x="label",
    data=df,
    order=df["label"].value_counts().index
)
plt.xticks(rotation=90)
plt.title("Crop Distribution")
plt.show()
```

## Visualization 2

### Temperature Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["temperature"],
    kde=True
)

plt.title("Temperature Distribution")

plt.show()
```

## Visualization 3

### Humidity Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(
    df["humidity"],
    kde=True
)
plt.title("Humidity Distribution")
plt.show()
```

## Visualization 4

### Rainfall Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(
    df["rainfall"],
    kde=True
)
plt.title("Rainfall Distribution")
plt.show()
```

## Visualization 5

## Soil pH Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(
    df["ph"],
    kde=True
)
plt.title("Soil pH Distribution")
plt.show()
```

## Visualization 6

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    x=df["N"]
)
plt.title("Nitrogen Distribution")
plt.show()
```

## Visualization 7

### Phosphorus Distribution

## Visualization 9
Correlation Heatmap

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df.select_dtypes(include=np.number).corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

## Visualization 10
Temperature vs Rainfall

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x="temperature",
    y="rainfall",
    data=df,
    hue="label",
    legend=False
)

plt.title("Temperature vs Rainfall")

plt.show()
```

## Visualization 11
Humidity vs Rainfall

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x="humidity",
    y="rainfall",
    data=df,
    hue="label",
    legend=False
)

plt.title("Humidity vs Rainfall")

plt.show()
```

## Visualization 12
Average Nitrogen by Crop

```python
plt.figure(figsize=(12,6))

df.groupby("label")["N"]\
.mean()\
.sort_values()\
.plot(kind="bar")

plt.title("Average Nitrogen by Crop")

plt.show()
```

## Visualization 13

```python
plt.figure(figsize=(12,6))

df.groupby("label")["P"]\
.mean()\
.sort_values()\
.plot(kind="bar")

plt.title("Average Phosphorus by Crop")

plt.show()
```

## Visualization 14
Average Potassium by Crop

```python
plt.figure(figsize=(12,6))

df.groupby("label")["K"]\
.mean()\
.sort_values()\
.plot(kind="bar")

plt.title("Average Potassium by Crop")

plt.show()
```
